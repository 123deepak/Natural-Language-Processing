```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
videos = pd.read_csv('C:\\Users\\deepak\\Downloads\\Youtube\\USvideos.csv')
```

```python
# Titles without removal of stop words
video_titles = " ".join(str(title) for title in videos.title)
#video_titles[0:1000]

titles = WordCloud(max_words=200).generate(video_titles)
plt.figure(figsize = (12, 8))
plt.imshow(titles, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
# Titles after removal of stop words
video_titles = " ".join(str(title) for title in videos.title)

# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["Official", "Video", "Music", "HD" , "Trailer", "Teaser", "Audio", "Lyric"])

titles = WordCloud(stopwords = stopwords, max_words=200).generate(video_titles)
plt.figure(figsize = (12, 8))
plt.imshow(titles, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
# Descriptions without removal of stop words
video_descriptions = " ".join(str(description) for description in videos.description)

descriptions = WordCloud(max_words=200, background_color="white").generate(video_descriptions)
plt.figure(figsize = (12, 8))
plt.imshow(descriptions, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
# Descriptions after removal of stop words
video_descriptions = " ".join(str(description) for description in videos.description)

# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["https", "youtube", "twitter", "facebook", "instagram", "channel", "watch", "snapchat", "bit" , "ly" "goo", "gl", "youtu", "nttps", "nFollow", "nhttp", "nsubscribe", "plus", "google", "use", "code", "Official", "Video", "Music", "HD" , "Trailer", "Teaser", "Audio", "Lyric", "IQid", "yt", "Subscribe", "playlist"])

descriptions = WordCloud(stopwords=stopwords, max_words=200).generate(video_descriptions)
plt.figure(figsize = (12, 8))
plt.imshow(descriptions, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
video_tags = " ".join(str(tag) for tag in videos.tags)
tags = WordCloud(max_words=200).generate(video_tags)
plt.figure(figsize = (12, 8))
plt.imshow(tags, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
